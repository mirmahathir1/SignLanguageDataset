```python
import moviepy.editor as mpy
from moviepy.video.fx.all import crop

clip = mpy.VideoFileClip("sample2.mp4")
(w, h) = clip.size
cropped_clip = crop(clip, width=w/3, height=(2*h)/3, x_center=w/6, y_center=h/2)

cropped_clip.write_videofile('cropped.mp4',audio_codec='aac')
```
